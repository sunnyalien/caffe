Usage of caffe pre-trained model 
==========================

Before you read this tutorial, make sure you install the ipython and ipython notebook. Both program will help you understand the pre-trained model effectively.

After you run "ipython notebook" in your terminal, you can run codes in this tutorial in cell mode. It is more convenient to see the result and easier to be more fimilair with the caffe pre-trained model.

<li>Classification</li>
--------------------------------------------

The code below are modified from:
http://nbviewer.ipython.org/github/BVLC/caffe/blob/master/examples/classification.ipynb#

(1) The code below sets up the needs for caffe pre-trained model

```python
import numpy as np # using numpy to do scientific computation
import matplotlib.pyplot as plt # using matplotlib to show the images
%matplotlib inline

# Make sure that caffe is on the python path:
caffe_root = '../'  # this file is expected to be in {caffe_root}/examples
import sys
sys.path.insert(0, caffe_root + 'python')

import caffe

# Set the right path to your model definition file, pretrained model weights,
# and the image you would like to classify.
MODEL_FILE = 'imagenet/imagenet_deploy.prototxt'
PRETRAINED = 'imagenet/caffe_reference_imagenet_model'
IMAGE_FILE = 'images/cat.jpg'
```

(2) Since this is an example for classification, we must know the image that will be classified.

```python
input_image = caffe.io.load_image(IMAGE_FILE)
plt.imshow(input_image)
```

(3) caffe.Classifier is used to generate the pre-trained net of hinton!!!

<b>Parameters of caffe.Classifier:<br></b>
a. mean<br>
b. input channel swapping: mapping RGB into the reference ImageNet model's BGR order<br>
c. raw scaling: multiplying the feature scale from the input [0,1] to the ImageNet model's [0,255].<br>
d. image_dimension

```python
net = caffe.Classifier(MODEL_FILE, PRETRAINED,
                       mean=np.load(caffe_root + 'python/caffe/imagenet/ilsvrc_2012_mean.npy'),
                       channel_swap=(2,1,0),
                       raw_scale=255,
                       image_dims=(256, 256))
```

(4) We will set the phase to test since we are doing testing, and will use CPU+GPU for the computation.If you want to use CPU only, check the origianal example code for more information!

```python
net.set_phase_test()
net.set_mode_gpu()
```

(5) Now, we will use the net generated by caffe.Classifier to predict the categories for the target image. 

<b>Parameters of predict:</b><br>
a. input image: can be one or more<br>
b. oversample: "True" for using center-cropped and corner-cropped target images as well as their mirrored versions to do the prediction. The overall images will become 10 pieces to oversample the input image.; "False" for only using the center-cropped version of target image to do the prediction."Note that this makes a single input, although if you inspect the model definition prototxt you'll see the network has a batch size of 10. The python wrapper handles batching and padding for you!"

```python
prediction = net.predict([input_image], oversample=True) 
print 'prediction shape:', prediction[0].shape
plt.plot(prediction[0])
print 'predicted class:', prediction[0].argmax()
```

"You can see that the prediction is 1000-dimensional, and is pretty sparse. The predicted class 281 is "Tabby cat." Our pretrained model uses the synset ID ordering of the classes, as listed in `../data/ilsvrc12/synset_words.txt` if you fetch the auxiliary imagenet data by `../data/ilsvrc12/get_ilsvrc_aux.sh`. If you look at the top indices that maximize the prediction score, they are cats, foxes, and other cute mammals. Not unreasonable predictions, right?" from caffe author<p>

<li>Feature Visualization</li>
---------------------------

The code below are modified from: http://nbviewer.ipython.org/github/BVLC/caffe/blob/master/examples/filter_visualization.ipynb

(1) Import required modules and set plotting parameters. The precedure is almost the same as classification part.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Make sure that caffe is on the python path:
caffe_root = '../'  # this file is expected to be in {caffe_root}/examples
import sys
sys.path.insert(0, caffe_root + 'python')

import caffe

MODEL_FILE = 'imagenet/imagenet_deploy.prototxt'
PRETRAINED = 'imagenet/caffe_reference_imagenet_model'
IMAGE_FILE = 'images/cat.jpg'

plt.rcParams['figure.figsize'] = (10, 10) # size of image
plt.rcParams['image.interpolation'] = 'nearest' # interpolation rule
plt.rcParams['image.cmap'] = 'gray' # color map: the gray-level image will be shown correctly; no effect on color image 
```

(2) Run the classifier to set up the network to test mode so we can observe the features of each layer.

```python
net = caffe.Classifier(MODEL_FILE, PRETRAINED, gpu=True, 
                       mean=np.load(caffe_root + 'python/caffe/imagenet/ilsvrc_2012_mean.npy'),
                       channel_swap=(2,1,0),
                       raw_scale=255,
                       image_dims=(256, 256))

input_image = caffe.io.load_image(IMAGE_FILE)
prediction = net.predict([input_image], oversample=True) 
```

(3) Before we examine the features, we need to know the structure of the network! There are more functions available for the network in pycaffe.py if you are interested in.

(3-1) structure of each layer

<b>Format:</b> 'layer name', (number of input, number of layer, size of each layer(x&y))<br>
Since the input is oversampled, the number of input change from 1 to 10.

```python
[(k, v.data.shape) for k, v in net.blobs.items()] # blob = binary large object 
```

(3-2) structure of each receptive field

<b>Format:</b> 'layer name', (number of receptive field in this layer, number of receptive field in previous layer(one GPU or two), size of each receptive field(x&y))

```python
[(k, v[0].data.shape) for k, v in net.params.items()]
```

(4) Here is the function to visualize the receptive field(filter) and the feature map.

<b>Format:</b>vis_square(data to be visulized, paddle or not?, paddle value???)

```python
# take an array of shape (n, height, width) or (n, height, width, channels)
#  and visualize each (height, width) thing in a grid of size approx. sqrt(n) by sqrt(n)
def vis_square(data, padsize=1, padval=0):
    data -= data.min()
    data /= data.max()
    
    # force the number of filters to be square
    n = int(np.ceil(np.sqrt(data.shape[0])))
    padding = ((0, n ** 2 - data.shape[0]), (0, padsize), (0, padsize)) + ((0, 0),) * (data.ndim - 3)
    data = np.pad(data, padding, mode='constant', constant_values=(padval, padval))
    
    # tile the filters into an image
    data = data.reshape((n, n) + data.shape[1:]).transpose((0, 2, 1, 3) + tuple(range(4, data.ndim + 1)))
    data = data.reshape((n * data.shape[1], n * data.shape[3]) + data.shape[4:])
    
    plt.imshow(data)
```

(4-1) Show the input image before visulize the receptive fields and featuer maps.

```python
# index four is the center crop
plt.imshow(net.deprocess('data', net.blobs['data'].data[4]))
```

(4-2) Visulize the receptive field.

```python
# the parameters are a list of [weights, biases]
filters = net.params['conv1'][0].data
vis_square(filters.transpose(0, 2, 3, 1))
```

(4-3) Visulize the feature map.

```python
feat = net.blobs['conv1'].data[4, :36]
vis_square(feat, padval=1)
```

(4-4) Moreover, we can show the values of nodes and the histogram of these values.

```python
feat = net.blobs['fc6'].data[4]
plt.subplot(2, 1, 1)
plt.plot(feat.flat)
plt.subplot(2, 1, 2)
_ = plt.hist(feat.flat[feat.flat > 0], bins=100)
```
